#Q1. Which state has the higest litracy rate in india in 2011?

```python
literacy=df.groupby('State')['Literacy'].max().sort_values(ascending=False).head(6)
print(literacy)
```

```python
literacy_df=literacy.reset_index()
fig=px.bar(literacy_df,x='State',y='Literacy',color_discrete_sequence=['blue'], text='Literacy')
fig.update_layout(title_x=0.5)
fig.update_layout(title_text="Literacy rate of states in india in 2011")
fig.show()
```

#Mizoram had the highest litracy rate in 2011 = **97.91%**

#Q2. Which district has the higest litracy rate in india in 2011?

```python
df.head(6)
```

```python
literacy_of_district=df.groupby('District')['Literacy'].max().sort_values(ascending=False).head(6)
```

```python
literacy_of_district_df=literacy_of_district.reset_index()
fig=px.bar(literacy_of_district_df,x='District',y='Literacy',color_discrete_sequence=['purple'], text='Literacy')
fig.update_layout(title_x=0.5)
fig.update_layout(title_text="Literacy rate of District of india in 2011")
fig.show()
```

#Serchhip(Mizoram) district had the highest literacy rate = **97.91%**

#Q3. Which state had the highest sex-ratio in 2011?

```python
df.head(6)
```

```python
sex_ratio=df.groupby('State')['Sex-Ratio'].max().sort_values(ascending=False).head(6)
```

```python
sex_ratio_df=sex_ratio.reset_index()
fig=px.bar(sex_ratio_df,x='State',y='Sex-Ratio',color_discrete_sequence=['blue'], text='Sex-Ratio')
fig.update_layout(title_x=0.5)
fig.update_layout(title_text='Sex ratio of states in india in 2011')
fig.show()
```

# Puducherry had the highest sex-ratio in 2011 = **1184 females per 1000 males**

#Q4. Which district had the highest sex-ratio in 2011?

```python
df.head(6)
```

```python
district_sex_ratio=df.groupby(['State','District'])['Sex-Ratio'].max().sort_values(ascending=False).head(6)
```

```python
district_sex_ratio_df=district_sex_ratio.reset_index()
fig=px.bar(district_sex_ratio_df,x='District',y='Sex-Ratio',color_discrete_sequence=['purple'],text='Sex-Ratio')
fig.update_layout(title_x=0.5)
fig.update_layout(title_text='Sex ratio of District in india in 2011')
fig.show()
```

#Mahe(puducherry) district had the higest sex-ratio in 2011 = **1184 females per 1000 males**
